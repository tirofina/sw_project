```python
mask = np.array(Image.open("picture.png"))
```

```python
f = open('one.txt','r', encoding='UTF8')
txt = f.readlines()
f.close()
```

```python
new_txt = []
for i in txt:
    new_txt.append(i.replace('\n',''))
```

```python
wc = WordCloud(font_path = 'C:/Users/j/AppData/Local/Microsoft/Windows/Fonts/netmarbleM.ttf',max_words=500, mask=mask, margin=10,
               random_state=1, contour_color='white', contour_width =3, max_font_size = 100).generate(str_txt)
wc.to_file("a_new_hope.png")
plt.axis("off")
plt.figure(figsize =(10,10))
plt.title("Default colors")
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
